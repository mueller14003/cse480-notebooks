### 1. Describe a decider to check whether the language of a given DFA $D$ is infinite.

A decider to check whether the language of a given DFA is infinite would have to keep track of the states it has visited and check for a loop in the DFA. If the DFA has a loop, then the language it produces is infinite.

### 2. Describe a decider to check whether a linear bounded automaton $L$ working on an input $w$ will go into an infinite loop.

To check whether an LBA will go into an infinite loop, we must take the size of its tape $S$, the size of the alphabet $A$, and the number of states of the LBA $Q$. The total number of possible configurations for the LBA is $Q \times S \times A^{S}$. If the machine does not halt within $Q \times S \times A^{S}$ steps, then by the pigeonhole principle, it is repeating itself and is stuck in an infinite loop.
